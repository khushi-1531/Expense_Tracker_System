```python
df['Total_Expense'] = df[expense_cols].sum(axis=1)
```

# STEP 12:Column-wise OUTLIER DETECTION (IQR)

```python
outlier_summary = {}

for col in numeric_df.columns:
    Q1 = numeric_df[col].quantile(0.25)
    Q3 = numeric_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = numeric_df[
        (numeric_df[col] < lower) | (numeric_df[col] > upper)
    ][col]

    outlier_summary[col] = outliers
```

```python
for col, values in outlier_summary.items():
    print(f"{col}: {len(values)} outliers")
```

```python
for col in numeric_df.columns:
    plt.figure(figsize=(4, 6))
    sns.boxplot(y=numeric_df[col])
    plt.title(f"Outliers in")
    plt.show()
```
